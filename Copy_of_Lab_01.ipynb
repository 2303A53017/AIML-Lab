```python
plt.scatter(weight,height)
plt.title("Relation between weight and height of patients")
plt.ylabel("Height in Meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight",fontsize=14,fontstyle='italic')

plt.bar(x=people,height=weight,width=0.6,color='red',edgecolor='k',alpha=0.6)

plt.xlabel("people",fontsize=15)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram opf patient weight",fontsize=10)
plt.xlabel("weight in kgs",fontsize=10)
plt.show()
```
